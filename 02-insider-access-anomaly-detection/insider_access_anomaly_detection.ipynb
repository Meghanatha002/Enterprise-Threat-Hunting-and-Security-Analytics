```python
# Meghanatha Reddy Pallem
# 
# Identify employee accesses to resources outside their own department

import pandas as pd

# Loading the CSV files
access = pd.read_csv("datalogUpdated.csv")
employees = pd.read_csv("employee_details.csv")
resources = pd.read_csv("dept_resources.csv")
```

```python
# Merging access log with employee details (by EmployeeID)
merged = pd.merge(access, employees, on="EmployeeID", how="left")

# Merging with resource ownership (by ResourceID)
merged = pd.merge(merged, resources, on="ResourceID", how="left")

print("Merged dataset sample:")
print(merged.head())
```

```python
# Keeping rows where employee's dept != resource's dept
outside = merged[merged["Department_x"] != merged["Department_y"]]

result_df = outside[[
    "EmployeeID",        
    "Name",
    "Department_x",
    "ResourceID",
    "Department_y",
    "Action",
    "Datetime",
    "Result"
]].copy()

# Renaming columns for clarity
result_df = result_df.rename(columns={
    "Department_x": "Employee_Dept",
    "Department_y": "Resource_Dept"
})

print("Filtered sample:")
print(result_df.head())
```

```python
# Dropping exact duplicate rows
result_df = result_df.drop_duplicates()

# Total count
print(f"Total outside-department access events: {len(result_df)}\n")

# Sample rows 
print("First 10 rows:")
print(result_df.head(10))

print("\nLast 10 rows:")
print(result_df.tail(10))
```

```python
# Droping duplicates by EmployeeID
result_df = result_df.drop_duplicates(subset=["EmployeeID"])

# Total count
print(f"Total unique employees who accessed outside their department: {len(result_df)}\n")

#Sample rows
print("First 10 rows:")
print(result_df.head(10))

print("\nLast 10 rows:")
print(result_df.tail(10))
```
